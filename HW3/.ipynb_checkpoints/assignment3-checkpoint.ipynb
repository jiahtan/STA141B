```python
import os
import pandas as pd
#Look through each row, see if Fresh1 is in there
def ExtractInfoRow(file, food, foodType):
    """Extracts the Fresh rows from the fruit Excel file 
    and place it into a row to be appended to a pandas dataframe.
    Input: Excel File, the food name, and foodtype
    Output: pandas row with the desired columns.
    """
    for i in range(len(file)):
        #If any of the strings contain Fresh1 or Fresh retrieve row
        if any(string in str(file.iloc[i,0]) for string in ["Fresh1", "Fresh"]):
            goalString = file.iloc[i,:]
            return(pd.DataFrame({'type': foodType, 
                    'food': food, 
                    'form': "Fresh1", 
                    'price_per_lb': goalString[1], 
                    'yield': goalString[3], 
                    'lb_per_cup': goalString[4], 
                    'price_per_cup': goalString[6]}, index=[0]))
        
def addToDF(foodType):
    """
    Function to create a pandas dataframe with row information 
    gained from ExtractInfoRow function.
    Input: Type of food, vegetable or fruit
    Output: Pandas dataframe with all the rows for that food added.
    """
    foodDF = pd.DataFrame()
    foodFolder = os.listdir('./assignment3_data/' + foodType)
    for file in foodFolder:    
        f = pd.read_excel('./assignment3_data/' + foodType + '/' + file)
        food, _ = file.split(".")
        line = ExtractInfoRow(f, food, foodType)
        foodDF = foodDF.append(line)
    return(foodDF)
```

```python
fruitDF = addToDF("fruit")
```
